```python
## querying dataset ##

# cara filter tahun
current_decade = df[df["year_id"] > 2010]
```

```python
# cara cari not null
games_with_notes = df[df["notes"].notnull()]
games_with_notes.shape
```

```python
# cara menggunakan group by
# mencari jumlah game berdasarkan tahun dan hasil 
# diatas tahun 2010 yang poinnya diatas 100

df [
    (df["year_id"] > 2010) &
    (df["pts"] > 100)
].groupby(["year_id", "game_result"])["game_id"].count()
```
