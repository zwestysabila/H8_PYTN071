# Analisa

## area plot

1. bagaimana trend kejahatan di london dari 2008 s.d. 2016 ?

```python
summary = df.groupby('year')['value'].sum()

summary.plot()
```

```python
summary.plot(kind='line')

plt.title('Crime Trend in London') # add a title to the area plot
plt.ylabel('Number of Crimes') # add y-label
plt.xlabel('Years') # add x-label

plt.show()
```

mengalami tingkat kejahatan tertinggi pada 2008, kemudian mengalami penurunan pada periode 2012 s.d. 2014

2. bagaimana tren 5 wilayah dengan tingkat kejahatan terbesar di London?

```python
total_b = df.groupby(['year','borough']).agg({'value':sum})

total_b_pivot = total_b.pivot_table('value', ['borough'], 'year') 

total_b_pivot['total'] = total_b_pivot.sum(axis=1)

total_b_pivot.sort_values(by='total', ascending=False, axis=0, inplace=True)

total_b_top5 = total_b_pivot.head(5)

total_b_top5 = total_b_top5.transpose()

total_b_top5.head(9).plot(kind='line', figsize=(15,8))

plt.title('Crime Trend of top 5 Areas') # add a title to the area plot
plt.ylabel('Number of Crimes') # add y-label
plt.xlabel('Years') # add x-label

plt.show()
```

```python
total_b_top5.head(9).plot(kind='area', 
                          alpha=0.25,
                          stacked=False,
                          figsize=(20, 10), 
                         )

plt.title('Crime Trend of top 5 Areas') # add a title to the area plot
plt.ylabel('Number of Crimes') # add y-label
plt.xlabel('Years') # add x-label

plt.show()
```

```python
tingkat kejahatan tertinggi di wilayah Westminster. sementara 4 lainnya memiliki tren hampir serupa
```

3. Bagaimana wilayah dengan tingkat kejahatan terendah di London?

```python
total_b_least5 = total_b_pivot.tail(5)

total_b_least5 = total_b_least5.transpose()

total_b_least5.head(9).plot(kind='area', 
                          alpha=0.25,
                          stacked=False,
                          figsize=(20, 10), 
                         )

plt.title('Crime Trend of least 5 Areas') # add a title to the area plot
plt.ylabel('Number of Crimes') # add y-label
plt.xlabel('Years') # add x-label

plt.show()
```

tingkat kejahatan di wilayah City of London sangat rendah sehingga hampir tidak terlihat pada grafik

## histogram

1. Karena tahun 2008 merupakah tahun dengan tingkat kejahatan tetinggi, bagaimana distribusi kejahatan 33 wilayah pada periode tersebut?

```python
histo = df.groupby(['year','borough']).agg({'value':sum})

histo_pivot = histo.pivot_table('value', ['borough'], 'year') 

count, bin_edges = np.histogram(histo_pivot['2008'])

histo_pivot['2008'].plot(kind='hist', figsize=(8,5), xticks=bin_edges)

plt.title('Histogram in 2008') # add a title to the area plot
plt.ylabel('Number of Borough') # add y-label
plt.xlabel('Number of Crimes') # add x-label

plt.show()
```

pada tahun 2008 terlihat 10 wilayah terdistribusi sekitar 19 ribu sampai 24 ribu jumlah kejahatan

## bar chart

1. Komparasi jumlah kejahatan di Westminster periode 2008 s.d. 2016

```python
temp = df.groupby(['year','borough']).agg({'value':sum})

temp = temp.pivot_table('value', ['borough'], 'year') 

# step 1 : get data
df_westminster = temp.loc['Westminster',years]

df_westminster
```

```python
# step 2 : plot data
df_westminster.plot(kind='bar', figsize=(10,6))

plt.title('Westminster crime from 2008 to 2016') # add a title to the area plot
plt.ylabel('Number of Crimes') # add y-label
plt.xlabel('Year') # add x-label

plt.show()
```

memuncak pada tahun 2012

## pie chart

1. cek jenis major crime yang terjadi dari tahun 2008 s.d. 2016

```python
df_major = df.groupby('major_category', axis=0).sum()
df_major
```

```python
explode_list = [0.15, 0, 0, 0, 0, 0, 0, 0.05, 0.05]

# Graded-Funtion Begin (~8 Lines)
df_major['value'].plot(kind='pie',
                      figsize=(15, 6),
                      autopct='%1.1f%%',
                      startangle=90,
                      shadow=True,
                      labels=None,         # turn off labels on pie chart
                      # the ratio between the center of each pie slice and the start of the text generated by autopct
                      pctdistance=1.12,
                      explode=explode_list  # 'explode'
                      )
# Graded-Funtion End

# scale the title up by 12% to match pctdistance
plt.title('Crime Major Category [2008 - 2016]', y=1.12)

plt.axis('equal')

# add legend
plt.legend(labels=df_major.index, loc='upper left')

plt.show()
```

sebagian besar adalah theft and handling

## box plot

1. Cek Box plot of Bexley Crime from 2008 - 2016

```python
df_bexley = temp.loc[['Bexley'], years]
df_bexley =  df_bexley.transpose()

sample_df_bexley = df_bexley.reset_index()

sample_df_bexley
```

```python
sample_df_bexley.plot(kind='box', figsize=(8, 6))

plt.title('Box plot of Bexley Crime from 2008 - 2016')
plt.ylabel('Number of Crimes')

plt.show()
```

## word clouds

```python
# import package and its set of stopwords
from wordcloud import WordCloud, STOPWORDS

print ('Wordcloud is installed and imported!')
```

```python
stopwords = set(STOPWORDS)
```

1. show major crime category

```python
# instantiate a word cloud object
majorcrime_wc = WordCloud(
    background_color='white',
    max_words=2000,
    stopwords=stopwords
)

# generate the word cloud
majorcrime_wc.generate(' '.join(df['major_category']))

```

```python
plt.imshow(majorcrime_wc, interpolation='bilinear')
# Graded-Funtion End

plt.axis('off')
plt.show()
```

2. show minor crime category

```python
# instantiate a word cloud object
minorcrime_wc = WordCloud(
    background_color='white',
    max_words=2000,
    stopwords=stopwords
)

# generate the word cloud
minorcrime_wc.generate(' '.join(df['minor_category']))

plt.imshow(minorcrime_wc, interpolation='bilinear')
# Graded-Funtion End

plt.axis('off')
plt.show()
```
